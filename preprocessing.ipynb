```python
# load dogs vs cats dataset, reshape and save to a new file
import sys
file_names=['/content/PetImages/Cat','/content/PetImages/Dog']
for folder in file_names:# traverse folder
    for filename in os.listdir(folder): # traverse file
        infilename = os.path.join(folder,filename)
        if not os.path.isfile(infilename): continue
        oldbase = os.path.splitext(filename)
        # convert .jpg files to .png
        newname = infilename.replace('.jpg', '.png')  
        output = os.rename(infilename, newname)
```

```python
# split the dataset into train and validation
import shutil
classes = ['Cat','Dog']
rootdir = '/content/PetImages'

for i in classes:
    os.makedirs(rootdir +'/train/' + str(i))
    os.makedirs(rootdir +'/validation/' + str(i))
    source = rootdir + '/' + str(i)
    allFileNames = os.listdir(source)
    np.random.shuffle(allFileNames)
    
    validation_ratio = 0.3
    train_FileNames, validation_FileNames = np.split(np.array(allFileNames),
                                                      [int(len(allFileNames)* (1 - validation_ratio))])

    train_FileNames = [source+'/'+ name for name in train_FileNames.tolist()]
    validation_FileNames = [source+'/' + name for name in validation_FileNames.tolist()]

    for name in train_FileNames:
      shutil.copy(name, rootdir +'/train/' + str(i))

    for name in validation_FileNames:
      shutil.copy(name, rootdir +'/validation/' + str(i))
```

```python
# clean the dataset from corrupted and non-image files
from PIL import Image

file_names=['/content/PetImages/train/Cat','/content/PetImages/train/Dog',
            '/content/PetImages/validation/Cat','/content/PetImages/validation/Dog']
for folder in file_names:
    for filename in os.listdir(folder):
        infilename = os.path.join(folder,filename)
        try:
            img = Image.open(infilename) # open the image file
            img.verify() # verify that the file is an image
        except (IOError, SyntaxError) as e:
            print('Corrupt file:', filename) # print out the names of corrupt files
            os.remove(infilename)
```
